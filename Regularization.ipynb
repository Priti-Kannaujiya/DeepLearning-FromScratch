```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_moons
```

```python
X,y=make_moons(n_samples=100,random_state=42,noise=0.25)
X
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
model=Sequential([
    Dense(128,activation='relu',input_dim=2),
    Dense(128,activation='relu'),
    Dense(1,activation='sigmoid')
])
```

```python
optimizer=Adam(learning_rate=0.01)
```

```python
model.compile(optimizer=optimizer,loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(X,y,epochs=500,validation_split=0.2)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X,y,clf=model,legend=2)
```

##***L2 Regularisation***

```python
model1=Sequential([
    Dense(128,activation='relu',input_dim=2,kernel_regularizer=tensorflow.keras.regularizers.l2(0.001)),
    Dense(128,activation='relu',kernel_regularizer=tensorflow.keras.regularizers.l2(0.001)),
    Dense(1,activation='sigmoid')
])
```

```python
optimizer1=Adam(learning_rate=0.01)
model1.compile(optimizer=optimizer1,loss='binary_crossentropy',metrics=['accuracy'])
```

```python
X
```

```python
history=model1.fit(X,y,epochs=500,validation_split=0.2)
```

```python
plot_decision_regions(X,y,clf=model1,legend=2)
```
