```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame([[8,8,1],[7,9,1],[6,10,0],[5,5,0]],columns=('cgpa','iq','placed'))
```

```python
def initialize_parameter(layer_dims):
    np.random.seed(3)
    parameter={}
    L= len(layer_dims)

    for l in range(1,L):
        parameter['W'+str(l)]=np.ones((layer_dims[l-1],layer_dims[l]))*0.1
        parameter['b'+str(l)]=np.zeros((layer_dims[l],1))
    return parameter
```

```python
parameter=initialize_parameter([2,2,1])
```

```python
def Linear_forward(A_prev,W,b):
    z=np.dot(W.T,A_prev)+b
    Z=1/(1+np.exp(-z))

    return Z
    
```

```python
def l_forward(X,parameter):
    A=X
    l=len(parameter)//2

    for l in range(1,l+1):
        A_prev=A
        Wl=parameter['W'+str(l)]
        bl=parameter['b'+str(l)]

        A=Linear_forward(A_prev,Wl,bl)
    return A_prev,A
```

```python
X=df[['cgpa','iq']].values[0].reshape(2,1)
y=df[['placed']].values[0][0]
A1,y_hat=l_forward(X,parameter)
```

```python
y_hat=y_hat[0][0]
print(A1)
print(y_hat)
```

```python
def update_parameter(parameter,y,y_hat,A1):
    parameter['W2'][0][0]=parameter['W2'][0][0] + (0.001*(y-y_hat)*A1[0][0])
    parameter['W2'][1][0]=parameter['W2'][1][0] + (0.001*(y-y_hat)*A1[1][0])
    parameter['b2'][0][0]=parameter['b2'][0][0] + (0.001*(y-y_hat))

    parameter['W1'][0][0]=parameter['W1'][0][0] +(0.001*(y-y_hat)*parameter['W2'][0][0]*A1[0][0]*(1-A1[0][0])*X[0][0])
    parameter['W1'][0][1]=parameter['W1'][0][1] +(0.001*(y-y_hat)*parameter['W2'][0][0]*A1[0][0]*(1-A1[0][0])*X[1][0])
    parameter['b1'][0][0]=parameter['b1'][0][0] + (0.001*(y-y_hat)*parameter['W2'][0][0]*A1[0][0]*(1-A1[0][0]))

    parameter['W1'][1][0]=parameter['W1'][1][0] +(0.001*(y-y_hat)*parameter['W2'][1][0]*A1[1][0]*(1-A1[1][0])*X[0][0])
    parameter['W1'][1][1]=parameter['W1'][1][1] +(0.001*(y-y_hat)*parameter['W2'][1][0]*A1[1][0]*(1-A1[1][0])*X[1][0])
    parameter['b1'][1][0]=parameter['b1'][1][0] + (0.001*(y-y_hat)*parameter['W2'][1][0]*A1[1][0]*(1-A1[1][0]))

    
```

```python
parameter=initialize_parameter([2,2,1])
epoch=50

for i in range(epoch):
    Loss=[]
    for j in range(df.shape[0]):
        X=df[['cgpa','iq']].values[j].reshape(2,1)
        y=df[['placed']].values[j][0]
        A1,y_hat=l_forward(X,parameter)
        y_hat=y_hat[0][0]

        update_parameter(parameter,y,y_hat,A1)
        Loss.append(-y*np.log(y_hat) - (1-y)*np.log(1-y_hat))

    print('Epoch - ',i+1,'Loss - ',np.array(Loss).mean())

    parameter
```
