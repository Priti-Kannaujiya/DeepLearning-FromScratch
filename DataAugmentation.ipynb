```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen=ImageDataGenerator(
    rotation_range=30,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',        # nearest: fills the images when there is shift of image with its nearest portion
    width_shift_range=0.2,      # fill_mode='reflect' reflection of image is shown along with the image
    height_shift_range=0.2      # constent refers to fill the image with black when there is shift in image
)
```

```python
img=image.img_to_array(img)
```

```python
img.shape
```

```python
input_batch=img.reshape(1,200,200,3)   # 1 represent the number of images that we are giving as input
```

```python
# flow is used when it is applied on single image and another one is flow from directory which is used when all images are processed during training time

import os

i=0
output_dir = '/content/augmented_images'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

for output in datagen.flow(input_batch, batch_size=1, save_to_dir=output_dir):
  i=i+1
  if i==10:
    break
```
